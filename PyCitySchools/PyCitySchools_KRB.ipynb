```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
school_name_count = len(school_data_complete_df["school_name"].unique())

# Calculate the total number of students
total_students = (school_data_complete_df["student_name"]).count()

# Calculate the total budget
total_budget = sum(school_data_df["budget"])

#Calculate the average math score
average_math_scores = (school_data_complete_df["math_score"]).mean()

#Calculate the average reading score
average_reading_scores = (school_data_complete_df["reading_score"]).mean()

# Calculate the percentage of students with a passing math score (70 or greater)
passing_math_count = (school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70]).count()["student_name"]
percentage_passing_math = ((passing_math_count / total_students)*100)

# Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading_count = (school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70]).count()["student_name"]
percentage_passing_reading = ((passing_reading_count / total_students)*100)

# Calculate the percentage of students who passed math and reading (% Overall Passing)
passing_reading_math_count = (school_data_complete_df.loc[(school_data_complete_df["reading_score"] >= 70)
                               &(school_data_complete_df["math_score"] >= 70)]).count()["student_name"]
percentage_passing_reading_math = ((passing_reading_math_count / total_students)*100)
```

```python
# Create a dataframe to hold the above results
distric_summary_df = pd.DataFrame({"Total Schools": [school_name_count],
                                                 "Total Students": [total_students],
                                                 "Total Budget": [total_budget],
                                                 "Average Math Score": [average_math_scores],
                                                  "Average Reading Score": [average_reading_scores],
                                                  "Students with Passing Math Score": [percentage_passing_math],
                                                   "Students with Passing Reading Score": [percentage_passing_reading],
                                                   "Students with Passing Math and Reading Score": [percentage_passing_reading_math]}) 
distric_summary_df.head()                 
```

```python
# Format all columns
distric_summary_df["Total Students"] = distric_summary_df["Total Students"].map("{:,}".format)
distric_summary_df["Total Budget"] = distric_summary_df["Total Budget"].map("${:,}".format)
distric_summary_df["Average Math Score"] = distric_summary_df["Average Math Score"].map("{:.2f}".format)
distric_summary_df["Average Reading Score"] = distric_summary_df["Average Reading Score"].map("{:.2f}".format)
distric_summary_df["Students with Passing Math Score"] = distric_summary_df["Students with Passing Math Score"].map("{:.4f}%".format)
distric_summary_df["Students with Passing Reading Score"] = distric_summary_df["Students with Passing Reading Score"].map("{:.4f}%".format)
distric_summary_df["Students with Passing Math and Reading Score"] = distric_summary_df["Students with Passing Math and Reading Score"].map("{:.4f}%".format)
distric_summary_df.head()
```

## School Summary

```python
# Using GroupBy to separate the data into fields according to individual school values
school_type = school_data_df.set_index(["school_name"])["type"]
student_count = school_data_complete_df.groupby(["school_name"]).count()["student_name"]
school_budget = school_data_complete_df.groupby(["school_name"]).mean()["budget"]
per_student_budget = (school_budget / student_count)
average_math_score = school_data_complete_df.groupby(["school_name"]).mean()["math_score"]
average_reading_score = school_data_complete_df.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Using GroupBy for Creating DataFrame for % Passing Math
passing_math_df = (school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70])
numbers_math_passing = passing_math_df.groupby(["school_name"]).count()["student_name"]
percentage_math_passing = (numbers_math_passing / student_count)*100
type(percentage_math_passing)
```

```python
# Using GroupBy for Creating DataFrame for % Passing Reading
passing_reading_df = (school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70])
numbers_reading_passing = passing_reading_df.groupby(["school_name"]).count()["student_name"]
percentage_reading_passing = (numbers_reading_passing / student_count)*100
percentage_reading_passing
```

```python
# Using GroupBy for Creating DataFrame for % Passing Reading & Math
passing_math_reading_df = (school_data_complete_df.loc[(school_data_complete_df["reading_score"] >= 70)
                               &(school_data_complete_df["math_score"] >= 70)])
numbers_math_reading_passing = passing_math_reading_df.groupby(["school_name"]).count()["student_name"]
percentage_math_reading_passing = (numbers_math_reading_passing / student_count)*100
percentage_math_reading_passing
```

```python
# Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({"School Type": school_type,
                                                  "Total Students": total_students,
                                                 "Total Students": student_count,
                                                 "Total School Budget": school_budget,
                                                  "Per Student Budget": per_student_budget,
                                                  "Average Math Score": average_math_score,
                                                   "Average Reading Score": average_reading_score,
                                                   "% Passing Math": percentage_math_passing,
                                                    "% Passing Reading": percentage_reading_passing,
                                                    "% Overall Passing": percentage_math_reading_passing})
school_summary_df
```

```python
# Clean up formatting of data
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,}".format)
school_summary_df
```
